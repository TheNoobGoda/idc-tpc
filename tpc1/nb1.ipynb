```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import statsmodels.api as sm
import numpy as np
```

Mean:
The mean, also known as the average, is a measure of central tendency in a set of data.
It is calculated by summing all the values in a dataset and dividing the sum by the total number of values.
Mathematically, the mean (μ or x̄) is represented as:
It provides an overall sense of the "typical" value in the dataset.

Variance:
Variance measures the spread or dispersion of a set of data points around the mean.
It quantifies how much individual data points deviate from the mean.
Mathematically, the variance (σ^2) is represented as:
A higher variance indicates greater variability in the data.

Standard Deviation:
The standard deviation is a statistic that represents the average amount of deviation or dispersion of data points from the mean.
It is the square root of the variance and is often denoted as σ (sigma).
Mathematically, the standard deviation (σ) is represented as:
A higher standard deviation means that the data points are more spread out from the mean.

```python
d = pd.read_csv("SAheart.csv")
mean = np.mean(d['obesity'])
variance = np.var(d['obesity'])
std = np.std(d['obesity'])

plt.hist(d['obesity'])
plt.axvline(mean, color='r', linestyle='dashed', linewidth=2, label='Mean')
plt.annotate(f'Variance: {variance:.2f}', xy=(0.76, 0.85), xycoords='axes fraction', fontsize=10)
plt.annotate(f'Standard Deviation: {std:.2f}', xy=(0.63, 0.75), xycoords='axes fraction', fontsize=10)

plt.xlabel('Frequency')
plt.ylabel('Obesity')
plt.legend()


plt.show()
```

Quartiles:
Quartiles are statistical measures that divide a data set into four equal parts, each containing 25% (one quarter) of the data points.
There are three quartiles that are commonly used:
First Quartile (Q1): This is the 25th percentile of the data, dividing the lowest 25% of the data.
Second Quartile (Q2): This is the median of the data, dividing the data into two halves, with 50% below and 50% above.
Third Quartile (Q3): This is the 75th percentile of the data, dividing the lowest 75% of the data.
Quartiles are used to understand the spread and distribution of data and are particularly useful for identifying outliers and assessing the skewness of a data set.

Mode:
The mode is a statistical measure that represents the most frequently occurring value in a data set.
Unlike the mean (average) and median (middle value), the mode is concerned with the frequency of values rather than their numerical values.
A data set can have multiple modes if more than one value occurs with the highest frequency, making it a bimodal or multimodal distribution.
The mode is especially useful for categorical and discrete data, but it can also be applied to continuous data, although continuous data sets may not have a distinct mode.
In some cases, a data set may have no mode if all values occur with the same frequency.

Covariance:
Covariance is a statistical measure that quantifies the degree to which two random variables change together.
It measures the relationship between two sets of data, indicating whether they tend to increase or decrease at the same time.
A positive covariance indicates that as one variable increases, the other tends to increase as well, while a negative covariance suggests that as one variable increases, the other tends to decrease.

Correlation:
Correlation is a standardized measure of the linear relationship between two continuous random variables.
Unlike covariance, correlation values are always between -1 and 1, making it easier to interpret.
It indicates both the strength and direction of the linear relationship between two variables.
A correlation of 1 represents a perfect positive linear relationship, -1 represents a perfect negative linear relationship, and 0 indicates no linear relationship between the two variables.
In summary, covariance measures the degree and direction of joint variability between two random variables, while correlation provides a standardized measure of the strength and direction of a linear relationship between two continuous variables, making it easier to compare and interpret across different datasets.
