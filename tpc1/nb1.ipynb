```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import statsmodels.api as sm
import numpy as np
```

Mean: 
The mean, commonly referred to as the average, is a measurement of a data set's central tendency.
It is determined by adding up all of the dataset's values, then dividing the total by the overall number of values.
It gives a general impression of the dataset's "typical" value.

Variance:
The spread or dispersion of a group of data points relative to the mean is measured by variance.
It measures how far the mean is deviated from by each unique data point.
Greater variability in the data is indicated by a bigger variance.

Standard Deviation: The standard deviation is a statistic that shows how typically far data points deviate from the mean.
It is frequently written as Ïƒ (sigma), which stands for the square root of variance.
Data points are further dispersed from the mean when the standard deviation is larger.

```python
d = pd.read_csv("SAheart.csv")
mean = np.mean(d['obesity'])
variance = np.var(d['obesity'])
std = np.std(d['obesity'])

plt.hist(d['obesity'])
plt.axvline(mean, color='r', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(mean-std, color='m', linestyle='dashed', linewidth=2, label='Std')
plt.axvline(mean+std, color='m', linestyle='dashed', linewidth=2)
plt.annotate(f'Variance: {variance:.2f}', xy=(0.76, 0.80), xycoords='axes fraction', fontsize=10)
plt.annotate(f'Standard Deviation: {std:.2f}', xy=(0.63, 0.75), xycoords='axes fraction', fontsize=10)

plt.xlabel('Frequency')
plt.ylabel('Obesity')
plt.legend()


plt.show()
```

In the graph aboce we can see a histogram of the data. The red lines representes the mean of the data and the purple lines are the mean +/- the standrd deviation.

Quartiles:
Statistical measurements known as quartiles divide a data set into four equal portions, each of which contains 25% (or one quarter) of the total number of data points.
Three quartiles are frequently employed:
First Quartile (Q1): This is the data's lowest 25%, or the data's 25th percentile.
Second Quartile (Q2): This is the median of the data, dividing the data into two halves, with 50% below and 50% above.
Third Quartile (Q3): This is the data's lowest 75% percentile, or the data's 75th percentile.
When analysing the spread and distribution of data, quartiles are particularly helpful in locating outliers and determining how skewed a data collection is.

Mode:
The most typical value in a data set is represented by the statistical metric known as the mode.
The mode is more interested in the frequency of values than their numerical values, in contrast to the mean (average) and median (middle value).
If more than one value occurs with the highest frequency in a data collection, it is said to have several modes, or to have a bimodal or multimodal distribution.
Although continuous data sets do not always have a clear mode, the mode can still be applied to them and is particularly helpful for categorical and discrete data.
If all values occur with the same frequency, a data collection may occasionally not have a mode.

```python
print(d.describe())

mode = scipy.stats.mode(d['obesity'])

plt.boxplot(d['obesity'])
plt.axhline(mode[0], color='r', linestyle='dashed', linewidth=1, label='Mode')

plt.legend()


plt.show()
```

Covariance:
Covariance is a statistical measure that quantifies the degree to which two random variables change together.
It measures the relationship between two sets of data, indicating whether they tend to increase or decrease at the same time.
A positive covariance indicates that as one variable increases, the other tends to increase as well, while a negative covariance suggests that as one variable increases, the other tends to decrease.

Correlation:
Correlation is a standardized measure of the linear relationship between two continuous random variables.
Unlike covariance, correlation values are always between -1 and 1, making it easier to interpret.
It indicates both the strength and direction of the linear relationship between two variables.
A correlation of 1 represents a perfect positive linear relationship, -1 represents a perfect negative linear relationship, and 0 indicates no linear relationship between the two variables.
In summary, covariance measures the degree and direction of joint variability between two random variables, while correlation provides a standardized measure of the strength and direction of a linear relationship between two continuous variables, making it easier to compare and interpret across different datasets.
